```python
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [86,80,79,83]
z = [20,30, 45,34]

c = ["red", "black", "blue", "green"]

plt.title("Language Popularity", fontsize=24)
plt.xlabel("Language")
plt.ylabel("Popularity of Languages in %")
plt.bar(x, y, color = c, width=0.5, label = "Popularity")
plt.bar(x,z, color = 'gray', width=0.5, label = "Popularity1")
 
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
max_t = [25,51,52,47,46,49,51]
min_t =[34,35,36,31,28,35,26]
avg_t = [44,45,46,41,39,38,42]

plt.title("Weather")
plt.xlabel("Days")
plt.ylabel("Temperature")

plt.plot(days,max_t, label="max")
plt.plot(days,min_t, label="min")
plt.plot(days,avg_t, label="avg")
plt.legend(loc='best')
plt.show()
```
