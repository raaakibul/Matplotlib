```python
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [86,80,79,83]

c = ["red", "black", "blue", "green"]

plt.title("Language Popularity")
plt.xlabel("Language")
plt.ylabel("Popularity of Languages in %")
plt.bar(x, y, color = c, width=0.5, align="edge")
 
plt.show()
```
